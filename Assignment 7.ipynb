```python
# 4 imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fb = pd.read_csv('fb.csv')
fb
```

```python
fb.isnull().sum()
```

```python
fb.dropna()
```

```python
fb.isnull().sum()
# Removing outliers!!
mask = fb['like'] > 1000
rows_to_drop = fb[mask].index
fb.drop(index=rows_to_drop, axis=0, inplace=True)
```

```python
# x = np.array([13,26,37,41,58,67,82])
# y = np.array([23,27,31,46,58,62,79])
# z = np.array([98,78,65,54,48,39,20])
```

```python
x = fb['Type'].values[:10]
x.reshape(-1,1)
y = fb['like'].values[:10]
y.reshape(-1,1)
z = fb['share'].values[:10]
z.reshape(-1,1)
```

## LINE PLOT

```python
plt.xlabel('Type of post')
plt.ylabel('Likes Count')
plt.title('Like Share Count')
plt.plot(x,y,color = 'blue', label = 'Likes', marker = 'o', mfc = 'red', ls = '--', ms = 7)
plt.plot(x,z,color = 'black', label = 'Share', marker = 'o', mfc = 'yellow', ls = '--', ms = 7)
plt.grid()
plt.legend()
```

## BAR GRAPH

```python
plt.xlabel('Type')
plt.ylabel('Likes')
plt.title('Likes Count', fontsize = 20)
plt.bar(x,y,width = 0.3)  # transparency of the bars
plt.grid()
```

## SUB PLOT

```python
plt.figure(figsize=(16,10))

plt.subplot(2,2,2)
plt.xlabel('Type')
plt.ylabel('Like Count')
plt.title('Like Count Plot')
plt.plot(x,y,color='blue',marker='o',mfc='red',ls='--',ms=7)

plt.subplot(2,2,1)
plt.xlabel('Type')
plt.ylabel('Share Count')
plt.title('Share Count Plot')
plt.plot(x,z,color='black',marker='o',mfc='yellow',ls='--',ms=7)
```

## SCATTER PLOT

```python
plt.xlabel('Type')
plt.ylabel('Like Count')
plt.scatter(x,y, marker = '>')
plt.grid()
```

## PIE CHART

```python
mycolor = ['red','green','yellow']
type = ['Photo','Status','Story']
data = [32,45,68]
plt.title('Pie Chart')
plt.pie(data, labels = type, autopct = '%.2f%%', startangle=90, labeldistance=1.1, radius = 0.8, colors = mycolor, shadow = True )
```

```python
plt.figure(figsize = (16,10))

plt.subplot(2,2,1)
plt.xlabel('Type of post')
plt.ylabel('Likes Count')
plt.title('Like Share Count')
plt.plot(x,y,color = 'blue', label = 'Likes', marker = 'o', mfc = 'red', ls = '--', ms = 7)
plt.plot(x,z,color = 'black', label = 'Share', marker = 'o', mfc = 'yellow', ls = '--', ms = 7)
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.xlabel('Type')
plt.ylabel('Likes')
plt.title('Likes Count', fontsize = 20)
plt.bar(x,y,width = 0.3)
plt.grid()

plt.subplot(2,2,3)
plt.xlabel('Type')
plt.ylabel('Like Count')
plt.scatter(x,y, marker = '>')
plt.grid()

plt.subplot(2,2,4)
mycolor = ['red','green','yellow']
type = ['Photo','Status','Story']
data = [32,45,68]
plt.title('Pie Chart')
plt.pie(data, labels = type, autopct = '%.2f%%', startangle=90, labeldistance=1.1, radius = 0.8,
        colors = mycolor, shadow = True );
```
