```python
A = air.dropna() # removes the entire row, if one of the value is null
A.shape
```

```python
A = air.fillna(method = 'pad') # replaces all missing values in dataframe from preceding row
```

## Replace null values using numpy

```python
import numpy as np
```

```python
A = air['Ozone'].replace(np.NaN,air['Ozone'].mean()) # replace NaN values with the mean of non-missing values
A.head()
```

```python
A = air['Ozone'].replace(np.NaN,air['Ozone'].median())  # with median of non-missing values
A.head()
```

## Impute the missing values

```python
from sklearn.impute import SimpleImputer as SI
```

```python
imp = SI(missing_values=np.NaN,strategy='mean') # calc mean for each col. and tranforms the dataframe
A = imp.fit_transform(air) # stores the result in NumPy array 'A'
A
```

```python
A = pd.DataFrame(A, columns=air.columns) # converts A into Data frame with the same col. names of air 
A.head()
```
