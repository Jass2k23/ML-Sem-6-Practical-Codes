```python
# 4 imports
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
link = "https://www.flipkart.com/apple-iphone-15-black-128-gb/product-reviews/itm6ac6485515ae4?pid=MOBGTAGPTB3VS24W&lid=LSTMOBGTAGPTB3VS24WVZNSC6&marketplace=FLIPKART"
```

```python
response = urlopen(link)
```

```python
soup = bs(response,'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('p',class_='_2NsDsF AwS1CA')
names
```

```python
address = soup.find_all('p',class_='MztJPv')
address
```

```python
title = soup.find_all('p',class_='z9E0IG')
title
```

```python
cust_name = []

for i in range(0,len(names)):
    cust_name.append(names[i].get_text())

cust_name
```

```python
cust_add = []

for i in range(len(address)):
    cust_add.append(address[i].get_text())

cust_add
```

```python
cust_title = []

for i in range(len(title)):
    cust_title.append(title[i].get_text())

cust_title
```

```python
df = pd.DataFrame()

df['Customer Name'] = cust_name
df['Reviews'] = cust_title
df['Customer Address'] = cust_add

df
```

```python
df.to_csv('output2.csv',index=False)
```
