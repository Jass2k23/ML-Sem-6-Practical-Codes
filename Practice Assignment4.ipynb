## 5. Reshaping

## Melting ( from wide to long)

```python
fil_transpose.melt() # for converting wide format dataset into long format dataset\
#When no specific columns are specified using the value_vars parameter, 
#all other columns are melted into two new columns: one containing the variable names and 
#the other containing the corresponding values.
```

```python
fil.melt(id_vars=['Category','Type'])
#This parameter specifies which columns to keep as identifier variables 
#(columns that remain unchanged). In this case, 'Category' and 'Type' columns are 
#kept as identifier variables, meaning they won't be melted but will remain as 
#columns in the resulting DataFrame.
```
